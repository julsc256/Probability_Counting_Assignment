## Question 1
___

There are 3 red balls, 1 green ball, and 2 yellow balls in an urn.
- a. What is the probability of drawing a green ball, P(Green)?
- b. What is the probability of drawing a red ball, P(Red)?
- c. If you draw two consecutive balls from the urn without replacement, what is the probability of drawing both yellow balls?
- d. Again, drawing two consecutive balls without replacement, what is the probability of not drawing the green ball?

### Solution

First I want define a function to find the probability of an event.
- To find the probability of an event is P(A) = events in A divided by the events in the whole set

```python
def p(x, t):
    probability = x / t
    return probability
```

### Part a.
___
There is 1 green ball in the urn and a total of 6 balls in the urn.  Therefore the events in A is 1 and the events in the whole set is 6.

```python
print(p(1,6))
print('Solution is: approx. 17%')
```

### Part b.
___
There are 3 red balls in the urn and a total of 6 balls in the urn.
Therefore the events in A is 3 and the events in the whole set is 6.

```python
print(p(3,6))
print('Solution is: 50%')
```

### Part c.
___
There are 2 yellow balls in the urn.  To draw two consecutive yellow balls, one can think of as the probability of yellow and yellow.  It's important to note that it states without replacement, which means that the total is changing with each draw.  Therefore, the probability of A is 2 with the whole set of 6 multiplied by the probability of B, which is 1 with the whole set of 5.

```python
print(p(2,6)*p(1,5))
print('Solution is: approx. 7%')
```

### Part d.
___
To think about this problem, one can think about the probability of not getting a green ball, which is 3 red balls and 2 yellow balls, for a total of 5 balls.  5 balls is the probability of A with a whole set of 6.  Then not getting a green again, without replacement, is the probability of B, which is 4 with a whole set of 5.  To find the probability of A and B, one needs to mulitiply them.

```python
# Getting two consecutive balls that are not green without replacement
print(p(5,6)*p(4,5))
print('Solution is: approx. 67%')
```

## Question 3
You are a high school gym teacher and the school currently has no sports teams.  You have 24 students in your gym class and want to see how well they play at a few different types of sports.
- a. First you want to try basketball.  You will select one team of five first and let's assume that order doesn't matter in basketball.
    - i. How many ways are there to select the frist team of five?
    - ii. After the first team is selected, how many ways are there to choose the second team?
- b. Now, try tennis.  Every player in tennis should feel comfortable playing close to the net and also along the baseline(meaning that order doesn't matter).
    - i. How many different teams of two can you make(without replacements)?
    - ii. You're getting ahead of yourself and imagining the championship picture.  How many ways could you line up the potential doubles teams(your answer from the previous part) in a photo?
- c. Lastly, you'd like to play some football.  Unlike basketball and tennis, players in football are usually in very specific roles.  Assuming that 11 players are on the field for each team at a given time, answer the following:
    - i. First, choose players for team A.  How many different teams could you make?
    - ii. Once team A is already built, how many ways could you make team B?

### Solution
___

## Part a.
___
First let's write functions for combinations and permutations.

```python
import math
def nCr(n,r):
    f = math.factorial
    return f(n)// f(r)//f(n-r)
```

```python
def nPr(n,r):
    f = math.factorial
    return f(n)// f(n-r)
# integer division //
```

i. Since order does not matter, we know that we are using combinations.  We have n = 24 and r = 5, because we have 24 students with 5 positions to be filled.

```python
print(nCr(24,5), 'ways')
```

ii. Once again, since order does not matter, combinations is being used again.  In this case, the 5 students are already in the first team, so we need to get the left over students.  24 minus 5 is 19, therefore, n = 19 and r = 5, still.

```python
print(nCr(19,5), 'ways')
```

## Part b.
___
i. Again, order is stated not to matter, so we will use combinations.  In this case, we only need 2 people for a team.  So n = 24 and r = 2.

```python
print(nCr(24,2), 'ways')
```

ii. In this case, we want to see how each of the types of ways the 276 can be formed, using 276!.

```python
print(math.factorial(276), 'ways')
```

## Part c.
___

i. This case states that order matters, so we know that we must use permutations.  For the first time, we have the 24 students to choose from to make the team of 11.  n = 24 and r = 11, but we will be using the permutations instead of combinations.

```python
print(nPr(24,11), 'teams')
```

ii. Since 11 students have already been choosen for the first team, we must subtract 24 and 11, to get 13 students left to make the second team.  Again, order matters, so we'll be using permutations.  In this case, n = 13 and r = 11.

```python
print(nPr(13,11), 'ways')
```
